# Notebook 08 — Final Model Comparison
### Comparing all ML and DL models on:
- Regression → WQI Prediction (R² Score, RMSE)
- Classification → Water Quality Category (Accuracy, F1 Score)

```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
with open('results_01_linear_logistic.pkl', 'rb') as f:
  res_01 = pickle.load(f)
with open('results_02_decision_tree.pkl', 'rb') as f:
  res_02 = pickle.load(f)
with open('results_03_random_forest.pkl', 'rb') as f:
  res_03 = pickle.load(f)
with open('results_04_xgboost.pkl', 'rb') as f:
  res_04 = pickle.load(f)
with open('results_05_knn.pkl', 'rb') as f:
  res_05 = pickle.load(f)
with open('results_06_svm.pkl', 'rb') as f:
  res_06 = pickle.load(f)
with open('results_07_deep_learning.pkl', 'rb') as f:
  res_07 = pickle.load(f)

print("All results loaded successfully")
```

```python
#Regression Comparison Table)

reg_summary = pd.DataFrame([
  {
      'Model': 'Linear Regression',
      'R2 Train': round(res_01['Linear Regression']['r2_train'], 4),
      'R2 Test':  round(res_01['Linear Regression']['r2_test'],  4),
      'RMSE':     round(res_01['Linear Regression']['rmse'],     2)
  },
  {
      'Model': 'Decision Tree',
      'R2 Train': round(res_02['Decision Tree Regression']['r2_train'], 4),
      'R2 Test':  round(res_02['Decision Tree Regression']['r2_test'],  4),
      'RMSE':     round(res_02['Decision Tree Regression']['rmse'],     2)
  },
  {
      'Model': 'Random Forest',
      'R2 Train': round(res_03['Random Forest Regression']['r2_train'], 4),
      'R2 Test':  round(res_03['Random Forest Regression']['r2_test'],  4),
      'RMSE':     round(res_03['Random Forest Regression']['rmse'],     2)
  },
  {
      'Model': 'XGBoost',
      'R2 Train': round(res_04['XGBoost Regression']['r2_train'], 4),
      'R2 Test':  round(res_04['XGBoost Regression']['r2_test'],  4),
      'RMSE':     round(res_04['XGBoost Regression']['rmse'],     2)
  },
  {
      'Model': 'KNN',
      'R2 Train': round(res_05['KNN Regression']['r2_train'], 4),
      'R2 Test':  round(res_05['KNN Regression']['r2_test'],  4),
      'RMSE':     round(res_05['KNN Regression']['rmse'],     2)
  },
  {
      'Model': 'SVM',
      'R2 Train': round(res_06['SVM Regression']['r2_train'], 4),
      'R2 Test':  round(res_06['SVM Regression']['r2_test'],  4),
      'RMSE':     round(res_06['SVM Regression']['rmse'],     2)
  },
  {
      'Model': 'Deep Learning',
      'R2 Train': round(res_07['DL Regression']['r2_train'], 4),
      'R2 Test':  round(res_07['DL Regression']['r2_test'],  4),
      'RMSE':     round(res_07['DL Regression']['rmse'],     2)
  }
])

print("=== Regression Results ===")
display(reg_summary)
```

```python
#(Classification Comparison Table)

clf_summary = pd.DataFrame([
  {
      'Model': 'Logistic Regression',
      'Acc Train': round(res_01['Logistic Regression']['acc_train'], 4),
      'Acc Test':  round(res_01['Logistic Regression']['acc_test'],  4),
      'F1 Test':   round(res_01['Logistic Regression']['f1_test'],   4)
  },
  {
      'Model': 'Decision Tree',
      'Acc Train': round(res_02['Decision Tree Classification']['acc_test'], 4),
      'Acc Test':  round(res_02['Decision Tree Classification']['acc_test'], 4),
      'F1 Test':   round(res_02['Decision Tree Classification']['f1_test'],  4)
  },
  {
      'Model': 'Random Forest',
      'Acc Train': round(res_03['Random Forest Classification']['acc_train'], 4),
      'Acc Test':  round(res_03['Random Forest Classification']['acc_test'],  4),
      'F1 Test':   round(res_03['Random Forest Classification']['f1_test'],   4)
  },
  {
      'Model': 'XGBoost',
      'Acc Train': round(res_04['XGBoost Classification']['acc_train'], 4),
      'Acc Test':  round(res_04['XGBoost Classification']['acc_test'],  4),
      'F1 Test':   round(res_04['XGBoost Classification']['f1_test'],   4)
  },
  {
      'Model': 'KNN',
      'Acc Train': round(res_05['KNN Classification']['acc_train'], 4),
      'Acc Test':  round(res_05['KNN Classification']['acc_test'],  4),
      'F1 Test':   round(res_05['KNN Classification']['f1_test'],   4)
  },
  {
      'Model': 'SVM',
      'Acc Train': round(res_06['SVM Classification']['acc_train'], 4),
      'Acc Test':  round(res_06['SVM Classification']['acc_test'],  4),
      'F1 Test':   round(res_06['SVM Classification']['f1_test'],   4)
  },
  {
      'Model': 'Deep Learning',
      'Acc Train': round(res_07['DL Classification']['acc_train'], 4),
      'Acc Test':  round(res_07['DL Classification']['acc_test'],  4),
      'F1 Test':   round(res_07['DL Classification']['f1_test'],   4)
  }
])

print("=== Classification Results ===")
display(clf_summary)
```

```python
#(Regression Bar Chart)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].barh(reg_summary['Model'], reg_summary['R2 Test'],
           color='steelblue', edgecolor='black')
axes[0].set_title('R² Test Score — All Models (Regression)')
axes[0].set_xlabel('R² Score')
axes[0].set_xlim(0, 1.05)
axes[0].axvline(x=1.0, color='red', linestyle='--', linewidth=1)
for i, v in enumerate(reg_summary['R2 Test']):
  axes[0].text(v + 0.01, i, str(v), va='center', fontsize=9)

axes[1].barh(reg_summary['Model'], reg_summary['RMSE'],
           color='coral', edgecolor='black')
axes[1].set_title('RMSE — All Models (Regression)')
axes[1].set_xlabel('RMSE (lower is better)')
for i, v in enumerate(reg_summary['RMSE']):
  axes[1].text(v + 0.5, i, str(v), va='center', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
#(Classification Bar Chart)
models = clf_summary['Model']
x = np.arange(len(models))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 5))
bars1 = ax.bar(x - width/2, clf_summary['Acc Test'],
             width=width, label='Accuracy', color='steelblue', edgecolor='black')
bars2 = ax.bar(x + width/2, clf_summary['F1 Test'],
             width=width, label='F1 Score', color='coral', edgecolor='black')

ax.set_xticks(x)
ax.set_xticklabels(models, rotation=25, ha='right')
ax.set_ylabel('Score')
ax.set_title('All Models — Classification Comparison (Accuracy & F1)')
ax.set_ylim(0.85, 1.02)
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)

for bar in bars1:
  ax.text(bar.get_x() + bar.get_width()/2,
          bar.get_height() + 0.002,
          str(round(bar.get_height(), 4)),
          ha='center', va='bottom', fontsize=8)

for bar in bars2:
  ax.text(bar.get_x() + bar.get_width()/2,
          bar.get_height() + 0.002,
          str(round(bar.get_height(), 4)),
          ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
#(Overfitting Analysis)

clf_summary['Overfit Gap'] = round(clf_summary['Acc Train'] - clf_summary['Acc Test'], 4)

fig, ax = plt.subplots(figsize=(10, 4))
colors = ['green' if g < 0.02 else 'orange' if g < 0.05 else 'red'
        for g in clf_summary['Overfit Gap']]

ax.bar(clf_summary['Model'], clf_summary['Overfit Gap'],
     color=colors, edgecolor='black')
ax.axhline(y=0.02, color='orange', linestyle='--', label='Mild overfit threshold')
ax.axhline(y=0.05, color='red',    linestyle='--', label='High overfit threshold')
ax.set_title('Overfitting Gap (Train Acc - Test Acc) — All Models')
ax.set_ylabel('Gap (lower is better)')
ax.set_xticklabels(clf_summary['Model'], rotation=25, ha='right')
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print("Green  = minimal overfitting  (gap < 0.02)")
print("Orange = mild overfitting     (gap 0.02–0.05)")
print("Red    = high overfitting     (gap > 0.05)")
```

```python
#(Best Model per Task)

best_reg = reg_summary.loc[reg_summary['R2 Test'].idxmax()]
best_clf = clf_summary.loc[clf_summary['F1 Test'].idxmax()]
least_overfit = clf_summary.loc[clf_summary['Overfit Gap'].idxmin()]

print("=== Best Model Summary ===")
print("Best Regression  (R2 Test) :", best_reg['Model'],
    "-> R2 =", best_reg['R2 Test'], "RMSE =", best_reg['RMSE'])
print("Best Classifier  (F1 Test) :", best_clf['Model'],
    "-> Acc =", best_clf['Acc Test'], "F1 =", best_clf['F1 Test'])
print("Least Overfit              :", least_overfit['Model'],
    "-> Gap =", least_overfit['Overfit Gap'])
```

## Final Observations

### Regression (WQI Prediction):
- Linear Regression scores R²=1.0 because WQI is mathematically derived from the features
- Tree-based models (Decision Tree, Random Forest, XGBoost) give more realistic scores
- Deep Learning approximates well but needs more epochs to match tree-based models

### Classification (Water Quality Category):
- All models perform strongly (>93%) — the classes are well separated in feature space
- Logistic Regression is surprisingly competitive — suggests linear boundaries work well
- Random Forest and XGBoost are most consistent — low overfitting + high accuracy
- Deep Learning is competitive but requires more tuning to beat ensemble methods
- SVM scores lower due to training on a sampled subset, not the full dataset

### Key Learnings:
- Simple models (Logistic Regression) can match complex ones when data is well structured
- Ensemble methods (Random Forest, XGBoost) are the most reliable choice for tabular data
- Deep Learning shines when data is large, unstructured, or has complex feature interactions
- Always compare Train vs Test scores — a perfect Train score is a red flag
